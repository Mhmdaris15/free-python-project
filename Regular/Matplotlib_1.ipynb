```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mpl
```

```python
plt.style.use('ggplot')
```

```python
fig, ax = plt.subplots()
plt.show()
```

```python
plt.plot([1,2,3,4,5,6],[5,10,15,20,25,30],color='r',marker=True)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
x = np.linspace(0,20,1000)
y = np.sin(x)
plt.plot(x,y,color='b')
plt.title('Function of sin(x)')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.xlim()
plt.ylim()
plt.show()
```

# b: blue
# g: green
# r: red
# c: cyan
# m: magenta
# y: yellow
# k: black
# yw: white

```python
x = np.linspace(0,20,1000)
y = np.sin(x)
plt.plot(x,y,color='g',linestyle='--')
plt.plot(x,np.cos(x),color='m')
plt.title('Function of sin(x)')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.show()
```

```python
x = np.linspace(0,12,1000)
y = np.sin(x)

plt.plot(x,np.tan(x), 'b^--', alpha=0.5, label='tan(x)') #alpha for opacity
plt.plot(x,np.cos(x), 'y*--', label='cos(x)') # Label is for legend
# You can combined color and linestyle to be 'r--' that mean is red color and -- linestyle
plt.legend()
plt.title('Function of sin(x)')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.show()
```
